```python
# imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
connection = "mysql+pymysql://root:root@localhost/sakila"
```

```python
engine = create_engine(connection)
```

```python
# confirm engine
engine
```

```python
q = """SELECT customer.first_name, customer.last_name, customer.email, address.address, city.city
FROM customer
JOIN address ON customer.address_id = address.address_id
JOIN city ON address.city_id = city.city_id
WHERE address.city_id=312
;"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT film.title, film.description, film.release_year, film.rating, film.special_features
FROM film
JOIN film_category ON film.film_id = film_category.film_id
JOIN category ON film_category.category_id = category.category_id
WHERE category.name = 'Comedy'
;"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT actor.last_name, film.title, film.release_year
FROM actor
JOIN film_actor ON actor.actor_id = film_actor.actor_id
JOIN film ON film.film_id = film_actor.film_id
WHERE actor.first_name='Johnny' AND actor.last_name='Lollobrigida'
;"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT actor.first_name, actor.last_name
FROM actor
JOIN film_actor ON actor.actor_id = film_actor.actor_id
JOIN film ON film.film_id = film_actor.film_id
WHERE film.title = 'Bingo Talented'
;"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT payment.customer_id, payment.amount
FROM payment
WHERE payment.amount > 
  (SELECT (2*AVG(payment.amount))
  FROM payment)
;"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT customer.first_name, customer.last_name, COUNT(payment.payment_id) AS 'num_payments'
FROM customer
JOIN payment ON customer.customer_id = payment.customer_id
GROUP BY customer.customer_id 
ORDER BY COUNT(payment.payment_id) DESC
LIMIT 5
;"""
pd.read_sql(q, engine)
```
